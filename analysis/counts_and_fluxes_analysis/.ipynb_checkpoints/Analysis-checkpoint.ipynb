```python
def getSingleTrace(specieName, replicate=0):
    
    t, counts = trajs[replicate].getNumberTrajectory(species = trajs[replicate].species(specieName))
    
    return t, counts
        
def getAverageTrace(specieName):
    
    for i in range(replicates):
        
        t, count = trajs[i].getNumberTrajectory(species = trajs[i].species(specieName))
        
        try:
            counts = counts + count
        except:
            counts = count
          
    counts = counts/replicates
        
    return t, counts
```

```python
def getSingleSpecTrace(specID, rep=1):
    
    metDF = traj.loc[ traj["Time"] == specID ]

    metTrace = []

    for columnName, columnData in metDF.iteritems():

        if columnData.values[0] != specID:

            metTrace.append(columnData.values[0])
            
    specData = np.array(metTrace)
    
    return specData
```

```python
def plotSingleSpec(specID):

    metDF = traj.loc[ traj["Time"] == specID ]

    metTrace = []

    for columnName, columnData in metDF.iteritems():

        if columnData.values[0] != specID:

            metTrace.append(columnData.values[0])

    #     print(columnData.values[0])

    time = np.linspace(start = 1, stop = totalTime, num = totalTime)
    
    plt.ylabel('Counts')
    plt.xlabel('Time (sec)')

    plt.plot(time, metTrace)
#     plt.xlim(0,60)
    plt.title(specID)
    plt.show()
```

```python
def plotSingleSpecConc(specID):

    NA = 6.022e23

    metTrace = getSingleSpecTrace(specID)

    time = np.linspace(start = 0, stop = totalTime, num = totalTime)
    
    volumeTrace = getSingleSpecTrace('Volume')
    
    volumeTrace = np.divide(volumeTrace, 1e21)
    
    concTrace = np.divide(metTrace, volumeTrace)
    
    concTrace = np.divide(concTrace, NA)
    
    concTrace = np.multiply(concTrace, 1000)
    
    plt.ylabel('Conc (mM)')
    plt.xlabel('Time (sec)')

    plt.plot(time, concTrace)
#     plt.xlim(0,60)
    plt.title(specID)
    plt.show()
```

```python
def getTotalRna(rnaID):
    
    rnaTrace = getSingleSpecTrace(rnaID)
    
    initRna = rnaTrace[0]
    
    locusNum = rnaID.split('_')[1]
    
    degID = 'DM_' + locusNum
    
    degTrace = getSingleSpecTrace(degID)
    
    rnapID = 'RPM_' + locusNum
    
    rnapTrace = getSingleSpecTrace(rnapID)
    
    rnaTotalTrace = [initRna]
    
    currentRna = int(initRna)
    
    for i in range(1,len(degTrace)):
        
#         if degTrace[i] > degTrace[i-1]:
            
#             currentRna = currentRna - (degTrace[i] - degTrace[i-1])
            
        if degTrace[i] > degTrace[i-1]:
            
            currentRna = currentRna - (degTrace[i] - degTrace[i-1])
            
            if currentRna<0:
                
                print(rnaID)
            
#             print('Minus RNA')
            
#         if rnapTrace[i] < rnapTrace[i-1]:
            
#             currentRna = currentRna + (rnapTrace[i-1] - rnapTrace[i])
            
        if rnapTrace[i] > rnapTrace[i-1]:
            
            currentRna = currentRna + (rnapTrace[i] - rnapTrace[i-1])
            
#             print('Plus RNA')
            
        rnaTotalTrace.append(currentRna)
        
    return rnaTotalTrace
```
