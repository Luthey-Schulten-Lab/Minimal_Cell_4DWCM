```python
logFile = '/home/zane/Models/mincell/mc4d/log_test.log'
```

```python
import numpy as np

from matplotlib import pyplot as plt
```

```python
with open(logFile) as file:
    logLines = [line.rstrip() for line in file]
```

```python
RDME_times = []
Ribo_times = []
CME_times = []
ODE_times = []
DNA_times = []
Write_times = []

for line in logLines:
    
    if line.startswith('Non-Hook RDME Time'):
        time = float(line.split(':')[1])
        RDME_times.append(time)

    if line.startswith('Ribo time'):
        time = float(line.split(':')[1])
        Ribo_times.append(time)

    if line.startswith('CME time'):
        time = float(line.split(':')[1])
        CME_times.append(time)

    if line.startswith('ODE time'):
        time = float(line.split(':')[1])
        ODE_times.append(time)

    if line.startswith('DNA time'):
        time = float(line.split(':')[1])
        DNA_times.append(time)

    if line.startswith('Communication and file write time'):
        time = float(line.split(':')[1])
        Write_times.append(time)

RDME = np.array(RDME_times)
Ribo = np.array(Ribo_times)
CME = np.array(CME_times)
ODE = np.array(ODE_times)
DNA = np.array(DNA_times)
Write = np.array(Write_times)
```

```python
times_hook = np.linspace(0,len(RDME_times)/80,num=len(RDME_times))
times_1s = np.linspace(0,len(CME_times),num=len(CME_times))
times_dna = np.linspace(0,len(DNA_times)*4,num=len(DNA_times))
```

```python
plt.plot(times_hook, RDME)
plt.ylim(0,0.5)
```
