```python
def getKcats(kinetic_params,rxn_name):
    
    tmpdf = kinetic_params[kinetic_params['Reaction Name']==rxn_name]
    kf = float(tmpdf[tmpdf['Parameter Type']=='Substrate Catalytic Rate Constant']['Value'].values[0])
    kr = float(tmpdf[tmpdf['Parameter Type']=='Product Catalytic Rate Constant']['Value'].values[0])
    
    return kf, kr

def getKMs(kinetic_params,rxn_name,species_list):
    
    KMs = {}
    for spc in species_list:
        tmpdf = kinetic_params[kinetic_params['Reaction Name']==rxn_name]
        tmpdf2 = tmpdf[tmpdf['Parameter Type']=='Michaelis Menten Constant']
        KMs[spc] = float(tmpdf2[tmpdf2['Related Species']=='M_'+spc]['Value'].values[0])
        
    return KMs

#Get Keq Values
def getKcatsKeq(kinetic_params,r):
    
    ignore_spcs = ['h_c','h2o_c','o2_c']
    subs = []
    prods = []
    nu_subs = {}
    nu_prods = {}
    
    for spc in r.reactants:
        if spc.id not in ignore_spcs:
            nu_subs[spc.id] = int(np.abs(r.get_coefficient(spc.id)))
            subs.append(spc.id)
    for spc in r.products:
        if spc.id not in ignore_spcs:
            nu_prods[spc.id] = int(np.abs(r.get_coefficient(spc.id)))
            prods.append(spc.id)
    
    KMs_subs = getKMs(kinetic_params,r.name,subs)
    KMs_prods = getKMs(kinetic_params,r.name,prods)
    
    kf, kr = getKcats(kinetic_params,r.name)
    
    if kr > 0.0:
        keq = kf/kr
        tmp_prod = 1.0
        for i in range(len(subs)):
            tmp_prod = tmp_prod*np.power(KMs_subs[subs[i]],nu_subs[subs[i]])
        tmp_prod = 1.0/tmp_prod
        for i in range(len(prods)):
            tmp_prod = tmp_prod*np.power(KMs_prods[prods[i]],nu_prods[prods[i]])
        keq = keq*tmp_prod
    else:
        keq = 1.0
    
    if r.reversibility == False:
        rev = 0
    else:
        rev = 1
    
    return kf, kr, keq, rev

def parse_MetRXN_KEGG(met_df,r):
    
    subs = []
    prods = []
    nu_subs = {}
    nu_prods = {}
    
    for spc in r.reactants:
        nu_subs[spc.id] = int(np.abs(r.get_coefficient(spc.id)))
        subs.append(spc.id)
    for spc in r.products:
        nu_prods[spc.id] = int(np.abs(r.get_coefficient(spc.id)))
        prods.append(spc.id)
            
#     print(subs)
#     print(prods)
            
    rxn_str = ''
    separator = ' + '
    
    sub_list = []
    for i in range(len(subs)):
        if subs[i] == 'h2o_c':
            tmp_KEGG_ID = 'C00001'
        elif subs[i] == 'o2_c':
            tmp_KEGG_ID = 'C00007'
        elif subs[i] == 'h_c':
            tmp_KEGG_ID = 'C00080'
        else:
            tmp_KEGG_ID = met_df.loc[met_df['Met ID']==subs[i],'KEGG ID'].iloc[0]
        tmp_KEGG_ID = 'kegg:'+tmp_KEGG_ID
        for j in range(nu_subs[subs[i]]):
            sub_list.append(tmp_KEGG_ID)
    
    subs_str = separator.join(sub_list)
#     print(subs_str)
    
    prod_list = []
    for i in range(len(prods)):
        if prods[i] == 'h2o_c':
            tmp_KEGG_ID = 'C00001'
        elif prods[i] == 'o2_c':
            tmp_KEGG_ID = 'C00007'
        elif prods[i] == 'h_c':
            tmp_KEGG_ID = 'C00080'
        else:
            tmp_KEGG_ID = met_df.loc[met_df['Met ID']==prods[i],'KEGG ID'].iloc[0]
        tmp_KEGG_ID = 'kegg:'+tmp_KEGG_ID
        for j in range(nu_prods[prods[i]]):
            prod_list.append(tmp_KEGG_ID)
    
    prods_str = separator.join(prod_list)
#     print(prods_str)
            
    rxn_str = subs_str + ' = ' + prods_str
    
    return rxn_str
```

```python

N_rxn = len(rxn_list)
print(N_rxn)
    
rxn_list.sort()
print(rxn_list)

# with open(data_file,'r') as f:
#     df = pandas.read_excel(f,sheet_name=datasheet,header=0)
df = pandas.read_excel(data_file,sheet_name=datasheet,header=0)
met_df = pandas.read_excel(data_file,sheet_name='Intracellular Metabolites',header=0)

kcat_forward = -np.ones((N_rxn),dtype=np.double)
kcat_reverse = -np.ones((N_rxn),dtype=np.double)
keq = -np.ones((N_rxn),dtype=np.double)
FBA_rev = {}

for i_rxn in range(N_rxn):
    rxn_df = df.loc[df['Reaction Name']==rxn_list[i_rxn]]
    kcat_forward[i_rxn] = rxn_df.loc[
        rxn_df['Parameter Type']=='Substrate Catalytic Rate Constant'
                                     ,'Value'].iloc[0]
    kcat_reverse[i_rxn] = rxn_df.loc[
        rxn_df['Parameter Type']=='Product Catalytic Rate Constant'
                                     ,'Value'].iloc[0]



rxn_rev = np.where(kcat_reverse > 1.0E-5, 1, 0).astype(np.int32)

# rxn_model = cobra.io.read_sbml_model('./iMB155_noUnqATP_lipdiomics_wPUNP5_noLPS_noPDHab.xml')
# rxn_model = cobra.io.read_sbml_model('./iMB155_noUnqATP_lipdiomics_wPUNP5_noNBtransport.xml')
rxn_model = cobra.io.read_sbml_model('./metabolic_networks/Syn3A_updated.xml')

keq_equilibrator = np.zeros((N_rxn),dtype=np.double)
balance_test = np.zeros((N_rxn),dtype=np.int32)
for i_rxn in range(N_rxn):
    
    rxn = rxn_model.reactions.get_by_id(rxn_list[i_rxn])
    
#     print(rxn.reactants)
    
    kcat_forward[i_rxn], kcat_reverse[i_rxn], keq[i_rxn], FBA_rev[rxn_list[i_rxn]] = getKcatsKeq(df,rxn)
    
    rxn_str = parse_MetRXN_KEGG(met_df,rxn)
    
    print(rxn_list[i_rxn] + ' --- ' + rxn_str)
    
    tmp_rxn = cc.parse_reaction_formula(rxn_str)
    if tmp_rxn.is_balanced() == False:
        print('The reaction is not balanced. ' + rxn_list[i_rxn])
        balance_test[i_rxn] = 1
    else:
#         print('The reaction is balanced.')
        dG_prime = cc.dg_prime(tmp_rxn)
        keq_equilibrator[i_rxn] = dG_prime.value.m_as("kJ/mol")
        print(keq_equilibrator[i_rxn])
        
    
print('Number unbalanced = '+str(np.sum(balance_test)))

dG_lim = 200

dG0_equilibrator = np.copy(keq_equilibrator)
dG0_kinetics = -RT*np.log(keq)

print(dG0_equilibrator.shape)
print(dG0_kinetics.shape)

remove_indices = np.argwhere(np.abs(keq_equilibrator)>dG_lim)
balance_test[remove_indices] = 1

keq_equilibrator = np.exp(-keq_equilibrator/(RT))
print(keq_equilibrator)
```

```python
#Get reaction stoichiometry, kcats, and KMs
def getKcatsKMs(kinetic_params,r):
    
    ignore_spcs = ['h_c','h2o_c','o2_c']
    subs = []
    prods = []
    nu_subs = {}
    nu_prods = {}
    
    for spc in r.reactants:
        if spc.id not in ignore_spcs:
            nu_subs[spc.id] = int(np.abs(r.get_coefficient(spc.id)))
            subs.append(spc.id)
    for spc in r.products:
        if spc.id not in ignore_spcs:
            nu_prods[spc.id] = int(np.abs(r.get_coefficient(spc.id)))
            prods.append(spc.id)
    
    KMs_subs = getKMs(kinetic_params,r.name,subs)
    KMs_prods = getKMs(kinetic_params,r.name,prods)
    
    kf, kr = getKcats(kinetic_params,r.name)
    
    
    return kf, kr, subs, prods, KMs_subs, KMs_prods, nu_subs, nu_prods

def getConcs(N_t,times,conc_df,subs,prods):
    
    concs = {}
    
    for i in range(len(subs)):
        
        concs[subs[i]] = np.zeros((N_t),dtype=np.double)
        
        for i_t in range(N_t):
            concs[subs[i]][i_t] = 1000*conc_df.loc[conc_df['RXN'] == subs[i], times[i_t]].values
            
    for i in range(len(prods)):
        
        concs[prods[i]] = np.zeros((N_t),dtype=np.double)
        
        for i_t in range(N_t):
            concs[prods[i]][i_t] = 1000*conc_df.loc[conc_df['RXN'] == prods[i], times[i_t]].values
    
    
    return concs

def getEnzymes(N_t,times,enzymes_df):
    
    enzymes = np.zeros((N_t),dtype=np.double)
    
    for i in range(len(subs)):
        
        concs[subs[i]] = np.zeros((N_t),dtype=np.double)
        
        for i_t in range(N_t):
            concs[subs[i]][i_t] = conc_df.loc[conc_df['RXN'] == subs[i], times[i_t]].values
            
    for i in range(len(prods)):
        
        concs[prods[i]] = np.zeros((N_t),dtype=np.double)
        
        for i_t in range(N_t):
            concs[prods[i]][i_t] = conc_df.loc[conc_df['RXN'] == prods[i], times[i_t]].values
    
    
    return enzymes
```

```python
df = pandas.read_excel(data_file,sheet_name=datasheet,header=0)

# rxn_model = cobra.io.read_sbml_model('./iMB155_noUnqATP_lipdiomics_wPUNP5_noNBtransport.xml')
rxn_model = cobra.io.read_sbml_model('./metabolic_networks/Syn3A_updated.xml')

# iterate over ordered list of reactions
N_t_ones = np.ones((N_t),dtype=np.double)
for i_rxn in range(N_order_rxn):
    
    # get reaction from SBML
    rxn = rxn_model.reactions.get_by_id(rxn_order_list[i_rxn])
    
    # get reaction parameters
    kf, kr, subs, prods, KMs_subs, KMs_prods, nu_subs, nu_prods = getKcatsKMs(df,rxn)
    
    # get concentrations of substrates and products
    concs = getConcs(N_t,times,conc_df,subs,prods)
    
    # convert normalize all concentrations by KMs
    for sub in subs:
        concs[sub] = concs[sub]/KMs_subs[sub]
    for prod in prods:
        concs[prod] = concs[prod]/KMs_prods[prod]
        
    forward_flux[:,i_rxn] = kf*forward_flux[:,i_rxn]
    temp_denom = np.ones((N_t),dtype=np.double)
    for sub in subs:
        forward_flux[:,i_rxn] = forward_flux[:,i_rxn]*np.power(concs[sub],nu_subs[sub])
        temp_denom = temp_denom*np.power(N_t_ones+concs[sub],nu_subs[sub])
    denom[:,i_rxn] = denom[:,i_rxn] + temp_denom
    
    reverse_flux[:,i_rxn] = kr*reverse_flux[:,i_rxn]
    temp_denom = np.ones((N_t),dtype=np.double)
    for prod in prods:
        reverse_flux[:,i_rxn] = reverse_flux[:,i_rxn]*np.power(concs[prod],nu_prods[prod])
        temp_denom = temp_denom*np.power(N_t_ones+concs[prod],nu_prods[prod])
    denom[:,i_rxn] = denom[:,i_rxn] + temp_denom
    
    print(rxn)
    locus = rxn.gene_reaction_rule
    locus = locus.split('_')[-1]
    locus = 'M_PTN_JCVISYN3A_'+locus+'_c'
    print(locus)
    
    for i_t in range(N_t):
        enzyme_concs[i_t,i_rxn] = 1000*enzyme_df.loc[enzyme_df['RXN'] == locus, times[i_t]].values
    
print(enzyme_concs)
```

```python
denom = np.reciprocal(denom)

print(denom)
forward_flux = forward_flux*denom
reverse_flux = reverse_flux*denom
net_flux = forward_flux - reverse_flux

forward_flux = forward_flux*enzyme_concs
reverse_flux = reverse_flux*enzyme_concs
net_flux = net_flux*enzyme_concs

print(net_flux)
```
